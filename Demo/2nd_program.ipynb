```python
# Write one generic function to find the nth highest in a list

def highest(num,rank):
  highest=0
  rank=int(rank)
  for value in num:
    if value>highest:
      highest=value
  
  for i in range(1,rank):
    a=0
    for j in num:
      if (j>a and j<highest):
        a=j
        high=a
    highest=high
  return high
def main():
  prices=[5,4,3,8,9,7]
  n=input('Rank: ')
  result=highest(prices,n)
  print('highest %s = %s'%(n,result))
main()
```
